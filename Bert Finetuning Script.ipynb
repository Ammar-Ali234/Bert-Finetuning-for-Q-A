```python
import torch
from torch.utils.data import DataLoader
```

```python
from sentence_transformers import SentenceTransformer, LoggingHandler
from sentence_transformers import models, util, datasets, evaluation, losses
import pandas as pd
from google.colab import drive
import pandas as pd
```

```python
#step1
from transformers import AutoModelForMaskedLM
bert_maskedML=AutoModelForMaskedLM.from_pretrained('bert-base-uncased')

from transformers import AutoTokenizer
bert_tokenizer=AutoTokenizer.from_pretrained("bert-base-uncased")


# Define your sentence transformer model using CLS pooling
model_name = "bert-base-uncased"
word_embedding_model = models.Transformer(model_name)
pooling_model = models.Pooling(word_embedding_model.get_word_embedding_dimension(), "cls")
model = SentenceTransformer(modules=[word_embedding_model, pooling_model])


```

```python
drive.mount('/content/drive')

# APath to train dataset in google drive
file_path = '/content/drive/My Drive/train_dataset.csv'

# Load the CSV file into a DataFrame
data = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to check if the file has been properly loaded
print(data.head())

#PROBLEM column as a list
train_sentences = data["PROBLEM"].tolist()

# Create the denoising auto-encoder dataset
train_dataset = datasets.DenoisingAutoEncoderDataset(train_sentences)

# DataLoader to batch your data
train_dataloader = DataLoader(train_dataset, batch_size=8, shuffle=True)

```

```python
import nltk
nltk.download('punkt')
```

```python

# Use the denoising auto-encoder loss
train_loss = losses.DenoisingAutoEncoderLoss(model, decoder_name_or_path=model_name,tie_encoder_decoder=True)
```

```python
# Call the fit method
model.fit(
    train_objectives=[(train_dataloader, train_loss)],
    epochs=6,
    weight_decay=0,
    scheduler="constantlr",
    optimizer_params={"lr": 3e-5},
    show_progress_bar=True,
)
```
